```python
def comparison_math_score_and_ethnic_group():
    """
    Plotting a Bar plot comparing math score of different ethnic group
    
    Parameters:
    math score and ethnic Column from DataFrame
    
    Returns the Bar Plot

    """
    plt.figure(dpi = 144)
    
    sns.barplot(data=df, x='race/ethnicity' , y='math score', hue='race/ethnicity')
    plt.xlabel('Group' , fontweight = 'bold')
    plt.ylabel('Math score' , fontweight = 'bold')
    plt.show()


comparison_math_score_and_ethnic_group()
```

```python
def correlation_of_variables():
    """
    Plots an heatmap showing the correlation between the scores in the dataset
    
    Parameters:
    The scores from the DataFrame
    
    Returns the heatmap plot

    """
    plt.figure(dpi = 144)

    sns.heatmap(data.corr(), annot=True, cmap='RdBu', vmin=-1, vmax=1)
    plt.title('Corelation')
    plt.show()

    
data = df.loc[:,['math score','reading score','writing score']]

correlation_of_variables()
```

```python
def kurtosis_of_data():
    """
    Gives kurtosis of Math , writing and reading score
    """
    kur = data.kurt()
    print(kur)  
  
    
data = df.loc[:,['math score','reading score','writing score']]

kurtosis_of_data()
```

```python
def skewness_of_data():
    """
    Gives skewness of math, reading, writing Score
    """
    skew = data.skew()
    print(skew)
   
    
data = df.loc[:,['math score','reading score','writing score']]

skewness_of_data()
```
