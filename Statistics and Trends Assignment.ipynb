```python
def comparison_math_score_and_ethnic_group():
    """
    Plotting a Bar plot comparing math score of different ethnic group
    
    Parameters:
    math score and ethnic Column from DataFrame
    
    Returns the Bar Plot
   
    """
    plt.figure(dpi = 144)
    sns.barplot(data=df, x='race/ethnicity' , y='math score', hue='race/ethnicity')
    plt.xlabel('Group' , fontweight = 'bold')
    plt.ylabel('Math score' , fontweight = 'bold')
    plt.show()



comparison_math_score_and_ethnic_group()
```

```python
def comparison_reading_score_parental_education():
    """
    Plotting a histogram comparing reading score of different parental level of education
    
    Parameters:
    reading score and parental level of education column from Data Set
    
    Returns the Bar Plot
   
    """
    plt.figure(dpi = 144)
    sns.barplot(data=df, x='parental level of education' , y='reading score', hue='parental level of education')
    plt.xlabel('Level Of Education' , fontweight = 'bold')
    plt.ylabel('Reading Score' , fontweight = 'bold')
    plt.xticks(rotation=90)
    plt.show()

comparison_reading_score_parental_education()
```

```python
def comparison_gender(gender):
    """
    Plotting a histogram comparing different gender
    
    Parameters:
    gender Column from DataFrame
    
    Returns the Histogram Plot
   
    """
    plt.figure(dpi = 144)
    sns.histplot(data=df, x='gender')
    plt.xlabel('Gender' , fontweight = 'bold')
    plt.ylabel('Frequency' , fontweight = 'bold')
    plt.title('Gender Comparison' , fontweight = 'bold') 
    plt.show()


comparison_gender(gender)

```

```python
def relation_btw_gender_and_math_score():
    """
    Plotting a box plot showing relation between gender and math score
    
    Parameters:
    gender and math column from the Data Set
    
    Returns the Box Plot
   
    """
    plt.figure(dpi=144)
    sns.boxplot(data=df, x='gender', y ='math score', hue='gender' )
    plt.xlabel('Gender' , fontweight = 'bold')
    plt.ylabel('Math Score' , fontweight = 'bold')
    plt.show()


relation_btw_gender_and_math_score()
```

```python
def relation_btw_lunch_nad_writing_score():
    """
    Plotting a violin plot showing the relation between type of lunch and writing score
    
    Parameters:
    lunch and writing score Column from DataFrame
    
    Returns the Violin Plot
   
    """
    plt.figure(dpi=144)
    colors=["#1f77b4", "#ff7f0e"]
    sns.violinplot(data=df, x='lunch', y ='writing score', hue='lunch' )
    plt.xlabel('Type of Lunch' , fontweight = 'bold')
    plt.ylabel('Writing Score' , fontweight = 'bold')
    plt.show()




relation_btw_lunch_nad_writing_score()
```
